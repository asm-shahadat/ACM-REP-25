```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Impurity
def misclassification(p):
    return 1-max(p, 1-p)

def gini(p):
    return 2*p*(1-p)

def cross_entropy(p):
    if (p == 0) or (p == 1):
        return 0
    else:
        return -p*np.log2(p) - (1-p)*np.log2(1-p)

probs = np.linspace(0, 1, 100)
y_m = [misclassification(x) for x in probs]
y_ce = [cross_entropy(x) for x in probs]
y_ce_norm = [x*0.5 for x in y_ce]
y_gini = [gini(x) for x in probs]

df = pd.DataFrame({
        'misclassification': y_m,
        'cross-entropy (scaled)': y_ce_norm,
        'gini': y_gini
    }, index=probs)

_ = df.plot()
_ = plt.xlabel('probability')
_ = plt.ylabel('impurity')
```
