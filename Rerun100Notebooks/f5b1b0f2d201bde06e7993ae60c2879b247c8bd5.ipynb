```python
def plothist(mu, sigma, Ngen):
    out = []
    for _ in range(Ngen):
        dist= np.random.normal(loc=mu, scale=sigma, size=1000)
        out.append(dist.mean())
    ttl = f'{Ngen} means of N({mu},{sigma})x1000'
    plt.hist(np.array(out))
    plt.title(ttl)
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2, 1)
plothist(10, 2, 50)
plt.subplot(1,2, 2)
plothist(10, 1, 1000)
plt.show()
```

The variable sampled 1000 times has a distribution closer to the normal distribution than the one sampled only 50 times, which supports the central limit theorem
